# RATING AGGREGATION

```python
zomato.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
```

```python
ratings=zomato.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"Rating count"})
ratings
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Aggregate rating',y='Rating count',hue='Rating color',data=ratings)
plt.xticks(rotation=90)
plt.show()
```

```python
ratings[ratings['Rating color']=='Orange']
```

# FREQUENCY DISTRIBUTION OF RATING COLOR IN DATA

```python
plt.figure(figsize=(10,6))
sns.countplot('Rating color',data=ratings)
plt.xticks(rotation=90)
plt.show()
```

```python
ratings[ratings['Rating color']=='Orange']
```

# WHICH COUNTRY HAS ONLINE DELIVERY

```python
z=zomato.groupby(['Country','Has Online delivery']).size().reset_index()
z
```

```python
plt.figure(figsize=(12,6))
plt.title('WHICH COUNTRY HAS ONLINE DELIVERY')
sns.barplot(z['Country'],z[0],hue='Has Online delivery',data=z)
plt.xticks(rotation=90)
plt.show()
```
